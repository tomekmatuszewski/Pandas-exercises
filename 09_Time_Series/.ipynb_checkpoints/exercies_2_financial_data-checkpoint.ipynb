### Step 7. Combine the four separate dataFrames into one combined dataFrame df that holds the information for all four stocks

```python
df_all_companies = pd.concat([df_apple, df_ibm, df_microsoft, df_tesla])
df_all_companies
```

### Step 8. Shift the stock column into the index (making it a multi-level index consisting of the ticker symbol and the date).

```python
df_all_companies['date'] = df_all_companies.index
df_all_companies

df_all_companies.set_index(['stock', 'date'], inplace=True)
```

```python
df_all_companies
```

### Step 7. Create a dataFrame called vol, with the volume values.

```python
vol = df_all_companies[['volume']]
vol
```

### Step 8. Aggregate the data of volume to weekly.
Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
df_all_companies.reset_index(inplace=True)
```

```python
df_all_companies.set_index(['stock','date'], inplace=True)
```

```python
vol = df_all_companies[['volume']]
vol
```

```python
vol_unstacked = vol.unstack(level=0)
vol_unstacked.head()
```

```python
vol_unstacked['date'] = vol_unstacked.index
vol_unstacked['date'] =  pd.to_datetime(vol_unstacked['date'])
vol_unstacked['year'] =  vol_unstacked['date'].dt.year
vol_unstacked['week'] =  vol_unstacked['date'].dt.isocalendar().week
```

```python
vol_unstacked.set_index(['year', 'week'], inplace=True)
```

```python
vol_unstacked.drop(columns='date', inplace=True)
```

```python
vol_unstacked
```

### Step 9. Find all the volume traded in the year of 2015

```python
vol_unstacked[vol_unstacked.index.isin(['2015'], level=0)].sum()
```
