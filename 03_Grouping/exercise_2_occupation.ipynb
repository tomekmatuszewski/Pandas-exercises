### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users_occupation = users.groupby(by='occupation')
users_occupation.groups.keys()

users_occupation[['age']].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
occupation_gender_count = pd.DataFrame(columns=['occupation', 'male', 'female'], )
```

```python
for occupation, occupation_data in users_occupation:
    count_male = occupation_data['gender'].where(occupation_data['gender']=='M').count()
    count_female = occupation_data['gender'].where(occupation_data['gender']=='F').count()
    
    occupation_gender_count=occupation_gender_count.append({'occupation': occupation, 'male': count_male, 'female': count_female}, ignore_index=True)

occupation_gender_count
```

```python
occupation_gender_count['male_ratio'] = (occupation_gender_count['male'] / (occupation_gender_count['male'] + occupation_gender_count['female']))*100
occupation_gender_count.sort_values('male_ratio', ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users_occupation['age'].agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
occupation_age_per_gender = pd.DataFrame(columns=['occupation', 'male_mean_age', 'female_mean_age'])
```

```python
for occupation, occupation_data in users_occupation:
    mean_age_male = occupation_data['age'].where(occupation_data['gender'] == 'M').mean()
    mean_age_female = occupation_data['age'].where(occupation_data['gender'] == 'F').mean()
    
    
    occupation_age_per_gender=occupation_age_per_gender.append({'occupation': occupation, 'male_mean_age': mean_age_male, 'female_mean_age':  mean_age_female}, ignore_index=True)
occupation_age_per_gender.drop_duplicates(['occupation'])
```

```python
users.groupby(['occupation', 'gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
users_occ_gender = users.groupby(by=['occupation', 'gender']).agg({'gender': 'count'})
users_occ_gender

occup_count = users.groupby(['occupation']).agg('count')
occup_count

occup_gender = users_occ_gender.div(occup_count) * 100

occup_gender[['gender']]
```
