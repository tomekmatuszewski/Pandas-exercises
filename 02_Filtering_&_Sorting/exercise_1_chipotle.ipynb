### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
converted_prices = [float(x[1:]) for x in chipo.item_price]

chipo.item_price = converted_prices
```

```python
chipo.head()
```

```python
chipo_drop_dupli = chipo.drop_duplicates(['item_name','quantity','choice_description'])

quantity_1 = chipo_drop_dupli['quantity'] == 1
chipo_per_item = chipo_drop_dupli[quantity_1]


chipo_filtered = chipo_per_item[chipo_per_item['item_price'] > 10.00]
chipo_filtered.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo.head(50)
chipo_without_duplicated = chipo.drop_duplicates(['item_name', 'choice_description'])
chipo_without_duplicated

chipo_without_duplicated['price_per_item'] = chipo_without_duplicated.item_price / chipo_without_duplicated.quantity
chipo_without_duplicated[chipo_without_duplicated['item_name'] == 'Nantucket Nectar']

# chipo_without_duplicated
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by='item_price', ascending=False).head(1)
```
