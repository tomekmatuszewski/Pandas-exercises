# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'

titanic = pd.read_csv(url)
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId', inplace=True)
```

```python
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
# sex = titanic.Sex.value_counts().plot(kind='pie')

pie, ax = plt.subplots(figsize=[10,6])

sex_occurrence = titanic.Sex.value_counts()

plt.pie(x=sex_occurrence, autopct="%.1f%%", pctdistance=0.5, explode=[0.05]*2, labe;s=['Male', 'Female'])
plt.title("sex-proportion", fontsize=14);
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.set(style='darkgrid')

fig, ax = plt.subplots(figsize=(15,5))

sns.scatterplot(data=titanic, x='Age', y='Fare', ax=ax, hue='Sex')
```

### Step 7. How many people survived?

```python
pie, ax = plt.subplots(figsize=[10,6])
total = titanic.Survived.count()
survived = titanic.Survived.value_counts()
autopct=lambda p: '{:.0f}'.format(p * total / 100)

plt.pie(x=survived, autopct=autopct, pctdistance=0.5, explode=[0.05]*2)
plt.title("titanic - survived", fontsize=14);
ax.legend(labels=['deaths', 'survived'])
plt.show()
```

```python
titanic.Survived.value_counts()
```

### Step 8. Create a histogram with the Fare payed

```python
sns.set(style='darkgrid')

fig, ax = plt.subplots(figsize=(15,7))

df = titanic.Fare.sort_values(ascending = False)
df

# create bins interval using numpy
binsVal = np.arange(0,600,10)
binsVal

# create the plot
plt.hist(df, bins = binsVal)

# Set the title and labels
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Payed Histrogram')

# show the plot
plt.show()
```
